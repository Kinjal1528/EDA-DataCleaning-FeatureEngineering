## FEATURE ENGINEERING 


```python
df['Date']=df['Date_of_Journey'].str.split('/').str[0]
df['Month']=df['Date_of_Journey'].str.split('/').str[1]
df['Year']=df['Date_of_Journey'].str.split('/').str[2]
```

```python
df['Date_of_Journey'].str.split('/').str[0]
```

```python
df.info()
```

We can see Date, Month and year are of object type but they must be of type integer so we will have to change it.  

```python
df["Date"]=df["Date"].astype(int)
df["Month"]=df["Month"].astype(int)
df["Year"]=df["Year"].astype(int)
```

```python
df.info()
```

```python
##DROP Date of journey 
df.drop("Date_of_Journey",axis=1,inplace=True)
```

```python
df.head()
```

```python
df["Arrival_hour"]=df["Arrival_Time"].str.split(':').str[0]
df["Arrival_min"]=df["Arrival_Time"].str.split(':').str[1]
```

```python
df.head(1)
```

Here the issue if we see is that for a few cases date is given with time, so when we split hours and minutes of arrival time there is date also displayed with arrival_min. so to solve this we will take the first part of string after one space, will be more understood with the code written below 

```python
df["Arrival_Time"]=df["Arrival_Time"].apply(lambda x:x.split(' ')[0])
```

```python
df["Arrival_hour"]=df["Arrival_Time"].str.split(':').str[0]
df["Arrival_min"]=df["Arrival_Time"].str.split(':').str[1]
```

```python
df.head(1)
```

Again convert these into integers(The arrival hours and min) 

```python
df['Arrival_hour']=df['Arrival_hour'].astype(int)
df['Arrival_min']=df['Arrival_min'].astype(int)
```

```python
df.drop("Arrival_Time",axis=1,inplace=True)
```

```python
df["Duration_hrs"]=df["Duration_hrs"].astype(int)
df["Duration_mins"]=df["Duration_mins"].astype(int)
```

We can see that the places where we have only mins, in duration hrs it is giving error, so let's resolve it 

```python
df["Duration_hrs"] = (df["Duration"].str.extract(r"(\d+)h")[0].fillna(0).astype(int))
df["Duration_mins"] = (df["Duration"].str.extract(r"(\d+)m")[0].fillna(0).astype(int))
```
